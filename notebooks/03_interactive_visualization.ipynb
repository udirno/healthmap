```python
def apply_filters(df, selected_genres, energy_range, danceability_range):
    """Apply multi-dimensional filtering to dataset."""
    filtered = df[
        (df['genre'].isin(selected_genres)) &
        (df['energy'].between(energy_range[0], energy_range[1])) &
        (df['danceability'].between(danceability_range[0], danceability_range[1]))
    ]
    return filtered

# Test filtering functionality
test_genres = ['electronic', 'classical']
test_energy = (0.3, 0.8)
test_danceability = (0.2, 0.7)

filtered_sample = apply_filters(interface.df, test_genres, test_energy, test_danceability)
print(f"Filtered result: {len(filtered_sample)} tracks from {filtered_sample['genre'].nunique()} genres")
```

```python
def display_track_details(df, track_indices):
    """Present detailed metadata for selected tracks."""
    
    if not track_indices:
        print("No tracks selected")
        return
    
    selected_tracks = df.iloc[track_indices]
    
    print(f"**Selected Tracks ({len(selected_tracks)}):**")
    for idx, track in selected_tracks.iterrows():
        print(f"\n• **{track['track_name']}** by {track['artist_name']}")
        print(f"  Genre: {track['genre']}")
        print(f"  Audio Profile: Energy={track['energy']:.2f}, Dance={track['danceability']:.2f}, Valence={track['valence']:.2f}")

# Demonstration with sample selection
sample_indices = [10, 150, 300, 450, 600]
display_track_details(interface.df, sample_indices)
```
